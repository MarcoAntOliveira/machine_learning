```python
import pandas as pd
pd.set_option("display.max_columns", 42)
dados = pd.read_csv("archive/2016-building-energy-benchmarking.csv")
```

```python
dados.head()
```

```python
# calculando a quantidade de dados faltantes e retornando o percentual
faltantes_percentual = (dados.isnull().sum()/len(dados['ENERGYSTARScore']))*100
print(faltantes_percentual)
```

```python
# substituindo os dados faltantes da coluna ENERGYSTARScore pela mediana 
dados['ENERGYSTARScore'].fillna(dados['ENERGYSTARScore'].median(), inplace= True)
```

```python
#substituindo os dados faltantes na coluna LargestPropertyUseType  pela media
dados['LargestPropertyUseTypeGFA'].fillna(dados['LargestPropertyUseTypeGFA'].median(), inplace= True)
```

```python
dados.dtypes
```

```python
dados['ThirdLargestPropertyUseTypeGFA'].fillna(dados['ThirdLargestPropertyUseTypeGFA'].median(), inplace= True)
```

```python
dados['SecondLargestPropertyUseTypeGFA'].fillna(dados['SecondLargestPropertyUseTypeGFA'].median(), inplace= True)
```

```python
faltantes_percentual = (dados.isnull().sum()/len(dados['ENERGYSTARScore']))*100
print(faltantes_percentual)
```
